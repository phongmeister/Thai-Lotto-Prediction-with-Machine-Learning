# Train Model

## Design model with requirements
1. Select 2-digits number
2. Select multiple numbers 
  - Number of lotto to be brought (In this case constant 5 number)
3. Using reward-penalty function
  - Reward penalty by prices
      - Win  = +2000 THB each
      - Lose = -Price each (Price = 80-100 THB) 
  - Reward-only when win
      - Win = +1
      - Loss = not count
4. Genetic
  - Each parameter in chromosome is random seed

```python
# You can open this one to try reward and penalty
reward_each = 2000
penalty_each = 80

def reward_function_with_penalty(actual_num, brought_list, penalty_mult=1, **kwargs):
    score = 0
    if actual_num in brought_list:
        score += sum(map(lambda x: x==actual_num, brought_list)) * reward_each
    # Include reward in list bacause each lotto cost penalty
    score -= len(brought_list) * (penalty_each * penalty_mult)
    return score

def reward_function_count(actual_num, brought_list, **kwargs):
    if actual_num in brought_list:
        return 1
    return 0

reward_function = reward_function_with_penalty
```

```python
def random_seed():
    return random.randrange(sys.maxsize)
```

```python
# Number of seed
NUM_SEED = 5

class Agent:

    def __init__(self, params=None):
        if params is None:
            params = self._random_params()
        self.params = params
    
    def _random_params(self):
        params = [random_seed() for i in range(NUM_SEED)]
        return params

    # time: time of prediction since start training....(since 01/09/2000)
    def agent_action(self, time):
        results = []
        
        for seed in self.params:
            num = 0
            # Set Seed
            random.seed(seed)
            # Simulate random time
            for i in range(time):
                num = random.randint(0, 99)
            results.append(num)
        
        # Reset seed
        random.seed(None)
        
        # Return brought number
        return results
    
    def __getstate__(self):
        return self.params
    
    def __setstate__(self, state):
        self.params = state
```

```python
# Using tournament selection 
#    [https://www.tutorialspoint.com/genetic_algorithms/genetic_algorithms_parent_selection.htm]
def select_parent(score_agent, k=5):
    # Pick sampling
    candidate = random.sample(score_agent, k)
    candidate.sort(key=lambda tup:tup[0], reverse=True)
    return candidate[0][1]
```

```python
# Create crossover class
def crossover(agent1, agent2):
    # Crossover return child
    parent_params = [agent1.params, agent2.params]
    
    # Shuffle random swap place between parent1 and parent2
    random.shuffle(parent_params)
    
    # Crossover only seed with index 1-4
    cross_idx = random.randint(1, NUM_SEED-1)
    seeds = parent_params[0][:cross_idx] + parent_params[1][cross_idx:]
        
    # Create child agent from crossover parameters
    child_agents = Agent(seeds)
    
    # Return children
    return child_agents
```

```python
  # Probability of mutation to occurs (5%) 
  # 01/09/2020 UPDATE TO 20%
  # 01/07/2022 UPDATE TO 25%
  def mutation(agent, mutation_prob=0.25):
      # Mutation return mutated agent
      params = agent.params
      for idx in range(len(params)):
          if random.random() < mutation_prob:
              # Mutate seed
              params[idx] = random_seed()

      # set params back
      agent.params = params
      return agent
```

### Set up simulate agent action using multiprocessing

```python
from multiprocessing import Pool

# This function has been optimized
def agent_execute_multiprocessing(args):
    params, datas, penalty_mult = args
    score = 0
    
    # Set empty list
    predicted_numbers = [[] for i in range(len(datas))]

    for seed in params:
        # Set seed
        random.seed(seed)
        
        for i in range(len(datas)):
            num = random.randint(0, 99)
            predicted_numbers[i].append(num)

    # Compute score in one time
    for actual, predicted in zip(datas, predicted_numbers):
        score += reward_function(actual, predicted, penalty_mult=penalty_mult)
    
    return score
```

```python
elite_pop = 0.05      # 5% of populations preserve as parent/breeder
crossover_pop = 0.6  # 60% of populations to control exploitation
alien_pop = 0.35      # 35% of populations to increase exploration

class LottoGeneticAlgorithm:

    def __init__(self, datas, num_agents=100):
        ''' Initialize object instance
        '''
        self.datas = datas
        self.num_agents=num_agents
        self.running_iter = 0
        self.agents = []
        self.scores = []
        self.average_elite_score = []
        self.best_elite_score = []
        self.average_total_score = []
        self.best_agent_overall = None
        self.best_score_overall = None
        self.init_agents()
        
    def init_agents(self):
        # Init agents with random parameters
        self.agents = [Agent() for i in range(self.num_agents)]
        
    def run(self, iteration=100, multiprocessing=False):
        ''' Run training GA for specified iteration/generation
        '''
        self.scores = []
        self.average_elite_score = []
        self.best_elite_score = []
        self.average_total_score = []
        start_time = time.time()

        print(f'Start running generic algorithm.')
        for i in range(iteration):
            print(f'Execute Iteration: {i+1}/{iteration}')
            self._run_iter(i, multiprocessing)
        
        # Compute last score for each agents
        if multiprocessing:
          self.scores = list(map(lambda tup: tup[0], self._run_agents_multiprocess(iteration)))
        else:
          self.scores = list(map(lambda tup: tup[0], self._run_agents(iteration)))
        print(f'Total time used for iteration[{iteration}]: {time.time()-start_time} seconds')
    
    def penalty_mult(self, iteration):
        # Increase penalty every 25 iteration for 200 iterations
        if iteration > 200:
            return 1
        return min(1, math.pow(2, int(iteration/25)-8))
    
    def _run_agents_multiprocess(self, iteration):
        scores = []
        agent_params = [(agent.params, self.datas, self.penalty_mult(iteration)) for agent in self.agents]
        with Pool() as pool:
            scores = pool.map(agent_execute_multiprocessing, agent_params)
        return list(zip(scores, self.agents))
        
    def _run_agents(self, iteration):
        scores = []
        for agent in self.agents:
            score = 0
            for i, num in enumerate(self.datas):
                numbers = agent.agent_action(i+1)
                score += reward_function(num, numbers, penalty_mult=self.penalty_mult(iteration))
            scores.append(score)
        return list(zip(scores, self.agents))

    def _run_iter(self, iteration, multiprocessing=False):
        ''' Method for running each iteration
        '''
        
        start_time = time.time()
        # Run each iteration
        score_agent = None
        if multiprocessing:
          score_agent = self._run_agents_multiprocess(iteration)
        else:
          score_agent = self._run_agents(iteration)
        
        print(f'Execution time used: {time.time()-start_time} seconds')
        
        # Do genetic algorithm
        score_agent.sort(key=lambda tup: tup[0], reverse=True)
        
        num_elite = int(elite_pop * self.num_agents)
        
        # Get first 10 best agent
        new_agents = list(map(lambda tup: tup[1], score_agent[:num_elite]))
        
        # Log best elite score
        self.average_elite_score.append(sum(map(lambda tup: tup[0], score_agent[:num_elite])) / num_elite)
        self.best_elite_score.append(score_agent[0][0])
        self.average_total_score.append(sum(map(lambda tup: tup[0], score_agent)) / len(score_agent))
        
        # Set best agent and best score for graph plotting later
        if self.best_agent_overall is None or self.best_score_overall < score_agent[0][0]:
            self.best_agent_overall = score_agent[0][1]
            self.best_score_overall = score_agent[0][0]
        
        # Show at least 1 elite
        for i in range(max(1, num_elite)):
            print(f'\tElite Score[{i+1}]: {score_agent[i][0]}\t{score_agent[i][1].params}')
        
        num_crossover = int(crossover_pop * self.num_agents)
        # Crossover pop
        for i in range(num_crossover):
            parent1 = select_parent(score_agent)
            # 01/09/2020 Select another parent from other than elite
            parent2 = select_parent(score_agent[num_elite:]) 
            child = crossover(parent1, parent2)
            new_agents.append(child)
            
        # Alien population
        num_alien = int(alien_pop * self.num_agents)
        for i in range(num_alien):
            new_agents.append(Agent())
        
        # 01/09/2020 Mutation only agent after 1st elite
        for idx in range(1, len(new_agents)):
            new_agents[idx] = mutation(new_agents[idx])
            
        self.agents = new_agents
```

# Evaluation

```python
def eval_model(model):
    # Plot learning graph
    fig = go.Figure(
        data=[
            go.Scatter(y=model.average_elite_score, x=list(range(len(model.average_elite_score))), name="Average Elite"),
            go.Scatter(y=model.best_elite_score, x=list(range(len(model.best_elite_score))), name="Best Elite"),
            go.Scatter(y=model.average_total_score, x=list(range(len(model.average_total_score))), name="Average Total"),
        ]
    )
    fig.update_layout(
        title="Genetic Algorithm - Fitness Score over generation",
        xaxis_title="Generation",
        yaxis_title="Fitness Score"
    )
    fig.show()

    # Pick best agent
    score_agent = list(zip(model.scores, model.agents))
    score_agent.sort(key=lambda tup: tup[0], reverse=True)
    best_score, best_agent = score_agent[0]
    print(f'Best agent parameters: {best_agent.params}')

    # Test
    count = 0
    for i, num in enumerate(model.datas):
      random.seed(None)
      should_buy = best_agent.agent_action(i+1)
      if num in should_buy:
        count+=1
    print(f'total won times is {count} out of {len(model.datas)}')
    print(f'The chance of winning lottery if playing with this seeds is {(count/len(model.datas))*100.0}')

    return best_agent
```

# *Pipeline*

```python
def pipeline(iteration=500, multiprocessing=True):
    data = load_data()
    data = clean_data(data)
    data = extract_features(data)
    model = LottoGeneticAlgorithm(data)
    # Run model
    model.run(iteration, multiprocessing)
    
    best_agent = eval_model(model)
    return best_agent
```
